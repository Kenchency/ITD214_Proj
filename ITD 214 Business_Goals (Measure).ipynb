```python
!pip install textblob
import pandas as pd
from textblob import TextBlob

# Load the CSV file
file_path = 'Lonely.csv'
data = pd.read_csv(file_path)

print(data.head())
```

```python
# Check the column names in your DataFrame
print(data.columns)
```

```python
# Assign column names
data.columns = ['lonely', 'feelings']
```

```python
print(data.columns)
```

```python
# Analyze sentiment for each sentence
data['Sentiment_Polarity'] = data['feelings'].apply(lambda x: TextBlob(x).sentiment.polarity)
data['Sentiment_Subjectivity'] = data['feelings'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Display the dataframe with sentiments
print(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the Sentiment Polarity
plt.figure(figsize=(10, 5))
sns.barplot(x=data.index, y=data['Sentiment_Polarity'], palette='PiYG')
plt.title('Sentiment Polarity of Sentences')
plt.xlabel('Sentence Index')
plt.ylabel('Sentiment Polarity')

# Rotate the x-axis labels to prevent overlap
plt.xticks(rotation=90, fontsize=8)

#show fewer x-axis labels
plt.xticks(ticks=range(0, len(data.index), 5), labels=data.index[::5], rotation=90, fontsize=8)

plt.tight_layout()  
plt.show()
```

```python
# Plotting the Sentiment Subjectivity
plt.figure(figsize=(10, 5))
sns.barplot(x=data.index, y=data['Sentiment_Subjectivity'], palette='PiYG')
plt.title('Sentiment Subjectivity of Sentences')
plt.xlabel('Sentence Index')
plt.ylabel('Sentiment Subjectivity')

# Rotate the x-axis labels to prevent overlap
plt.xticks(rotation=90, fontsize=8)

#show fewer x-axis labels
plt.xticks(ticks=range(0, len(data.index), 5), labels=data.index[::5], rotation=90, fontsize=8)

plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt

# Analyze sentiment for each sentence
data['Sentiment_Polarity'] = data['feelings'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Categorize the sentiments
def categorize_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

data['Sentiment_Category'] = data['Sentiment_Polarity'].apply(categorize_sentiment)

# Calculate percentage of each sentiment category
sentiment_counts = data['Sentiment_Category'].value_counts(normalize=True) * 100

# Display the percentages
print(sentiment_counts)

# Plotting the Sentiment Distribution as a Pie Chart
plt.figure(figsize=(8, 8))
colors = ['#FFFFE0', '#98FF98', '#FF69B4']
sentiment_counts.plot(kind='pie', colors=colors, autopct='%1.1f%%', startangle=140, labels=sentiment_counts.index, wedgeprops={'edgecolor': 'black'})
plt.title('Sentiment Distribution')
plt.ylabel('')  
plt.show()
```
